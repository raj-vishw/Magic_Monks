```python
from ultralytics import YOLO

models = {
    "fire": YOLO("fire_detection.pt"),
    "gun": YOLO("guns_yolov8n.pt"),
    # "weapon": YOLO("weapon detection.pt"),
    "knife": YOLO("weight_knief.pt")
}

img = "image/gun_shot.jpg"

for name, model in models.items():
    r = model(img)[0]
    if len(r.boxes) > 0:
        print(f"{name.upper()} DETECTED")
    else:
        print(f"{name.upper()} NOT detected")
```
